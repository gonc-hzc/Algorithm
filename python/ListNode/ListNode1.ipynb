链表由若干链节点通过 next 指针依次连接而成

- 链节点类(ListNode)：包含成员变量val (存储数据元素的值) 和 指向下一个节点的指针 next
- 链表类(LinkedList)：包含一个链接点变量head，用于表示链表的头节点

创建空链表时，只需将头节点 head 设为空指针 ，在 Python 中可用 None表示

## 链节点类

```python
class ListNode:
    def __init__(self , val = 0 , next = None):
        self.val = val    # 节点的值
        self.next = next  # 指向下一个节点

#不传任何参数（默认值）
node1 = ListNode()
print(node1.val)
print(node1.next)

#只传val，next用默认值
node2 = ListNode(5)
print(node2.val)
print(node2.next)

#传val和next（构建节点间的连接）
node3 = ListNode(8,node2)
print(node3.val)   # 输出：8
print(node3.next)  # 输出：<__main__.ListNode object at 0x...>（node2的内存地址）
print(node3.next.val)  # 输出：5（通过node3访问node2的值）
```

## 链表类

```python
class LinkList:
    def __init__(self):
        self.head = None
```

依次生成链表节点，并将它们顺序连接起来，构成完整的链表

1. 取出线性表的第 1 个元素，创建链表头节点。
2. 依次遍历剩余元素，每获取一个数据元素，就新建一个节点，并将其连接到当前链表的尾部。
3. 所有元素插入完成后，返回头节点

```python
    # 根据 data 列表初始化一个新链表
    def create(self , data):
        if not data: #等价于 if bool(data) is False:，即判断 data 转换为布尔值后是否为 False
            return  #如果输入数据为空，直接返回，不创建链表
        
        #创建头节点 并将 head 指向头节点
        self.head = ListNode(data[0])
        cur = self.head #cur指向当前链表的尾节点

        # 依次遍历 data 中剩余的元素，逐个创建新节点并连接到链表尾部
        for i in range(1,len(data)):
            node = ListNode(data[i])  #创建新节点
            cur.next = node           #将新的节点连接到当前的尾节点
            cur = cur.next             # cur 指向新的尾节点，准备连接下一个节点
```

创建链表的操作需要遍历所有数据元素，时间复杂度为 O(n)，其中 n 为线性表的长度。

## 链表长度
链表长度：通过一个指针变量cur沿着链表的next指针逐个遍历节点，并用计数器count统计节点数量，求出链表长度

1. 令指针cur指向链表头节点（第一个节点）
2. 沿着next指针遍历列表，每访问一个节点，计数器count+1
3. 当cur变为None（遍历到链表末尾）时，返回count的值

```python
    def length(self):
        count = 0           # 初始化计数器，记录节点个数
        cur = self.head     # 从链表头节点开始遍历

        while cur:          # 只要当前节点不为 None，就继续遍历
            count += 1
            cur = cur.next  # 指针后移，指向下一个节点
        return count
```

## 查找节点
从头节点head开始，遍历链表，查找值为val的节点。 找到：返回该节点，没找到：返回None

需要遍历链表的所有节点，因此其时间复杂度为O(n)

```python
    def find(self , val):
        cur = self.head         # 从链表头节点开始遍历
        while cur:              # 只要当前节点不为 None，就继续遍历
            if val == cur.val:  # 如果当前节点的值等于目标值，查找成功
                return cur      # 返回当前节点
            cur = cur.next      # 指针后移，指向下一个节点
        return None             #遍历完链表都没找到目标值，返回None
```

## 插入节点
在链表的第 i 个位置前插入一个值为 val 的新节点
1. 指针cur初始指向头节点，计数器count初始值0
2. 沿着链表的next指针遍历，cur每指向一个节点，count+1
3. 当count = index -1 ， cur正好指向第index - 1个节点（新节点的前一个节点），此时停止遍历
4. 创建一个新节点node 值为val
5. 将node.next指向cur.next
6. 将cur.next指向node ，完成插入操作

注意：如果index=1，即在头节点前插入，需要特殊处理（如使用虚拟头节点或单独判断）

操作需要将指针 cur 从链表头部遍历到第 i 个节点的前一个位置，平均时间复杂度为 O(n)


```python
# 插入节点
def insertInside(self , index , val):
    # 头部插入节点
    if index == 1:
        node = ListNode(val)
        node.next = self.head
        self.head = node

    #在index的位置插入节点
    count = 0
    cur = self.head
    while cur and count < index - 1:
        cur = cur.next
        count += 1

        #如果遍历完链表仍然没找到index的前一个节点，说明index越界了，插入失败
        if not cur:
            return 'Error'
        
        #尾部插入
        node = ListNode(val)
        if cur.next is None:
            cur.next = node
        #正常在中间插入节点的情况
        else:
            node.next = cur.next
            cur.next = node
```

## 修改节点
1. 指针cur初始指向头节点，计数器count初始值0
2. 沿着链表的next指针遍历，cur每指向一个节点，count+1
3. 当count = index的时候 ， cur指向第i个节点 ， 停止遍历
4. 直接将cur指向的节点的值改为val

将链表中第 i 个节点的值修改为 val，需要从链表头节点出发，遍历到第 i 个节点，然后进行赋值操作。由于遍历链表的时间复杂度为 O(n)

```python
def change(self , index , val):
    count = 0
    cur = self.head

    while cur and count < index:
        count += 1
        cur = cur.next

    #如果cur为空，说明index越界
    if not cur:
        return 'Error'
    
    #修改第index个节点的值为val
    cur.val = val
```

```python
LinkList.change = change
```
