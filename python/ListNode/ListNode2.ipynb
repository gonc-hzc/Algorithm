## 链表排序

| 适用性 | 排序方法 | 说明 |
|  :---:  |  :---:  |  :---:  |
| 适合|冒泡排序 、 选择排序 、入排序、归并排序、快速排序 | 适合链表结构|
| 需要额外空间 | 计数排序 、 桶排序 、 奇数排序 | 需要辅助数组，但是逻辑合适 |
|不合适|希尔排序|依赖随机访问，无法高效实现|
|不推荐|堆排序|完全二叉树结构用链表存储效率低|

## 冒泡排序

通过相邻节点比较和交换，将最大值逐步「冒泡」到链表末尾。
1. 外层循环：控制排序轮数，每轮将当前最大值「冒泡」到末尾
2. 内层循环：比较相邻节点，必要时交换值
3. 尾指针优化：每轮结束后，末尾已排序部分不再参与比较

外层：假设有 n 个数 ，每循环一轮就会排序一个数字，排到倒数第二个数字最后一个数字顺序也是正确的。 需要n - 1轮 
内层： n - i - 1 轮
