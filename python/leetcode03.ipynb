无重复字符的最长子串

给定一个字符串s,找到其中不含有重复字符的最长子串长度

方法：滑动窗口
1. right指针每次循环向右扩展窗口
2. 当遇到重复字符串时，left指针向右收缩窗口
3. 用set集合储存窗口内的字符



```python
class Solution:
    def lengthofsubstring(self , s:str) -> int:
        char_set = set()  #创建一个集合，用于存储窗口中的字符
        left = 0  #左指针 初始化位置为0
        max_length = 0

        for right in range(len(s)):
            #如果s[right]在char_set里已经出现过了，那么收缩左边界直至重复元素在窗口中被剔除
            while s[right] in char_set:
                char_set.remove(s[left])
                left += 1

            #将当前字符添加进窗口
            char_set.add(s[right])

            #记录最大子串长度
            max_length = max(max_length , right - left + 1)

        return max_length
```

```python
sol = Solution()

test_str = 'abcdeff'

res = sol.lengthofsubstring(test_str)
res = sol.lengthofsubstring('abca')

print(res)
```
