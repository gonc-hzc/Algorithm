```python
#两数相加
#链表：数组是一段连续的内存，链表是分散的
#链表是由一个一个节点串起来，每个节点在内存中独立
#head头节点：链表的入口，存放指向第一个节点的地址
#tail尾节点：最后一个节点，指向none表示结束
#中间每个节点都是 （数据 + 下一个节点地址 ）的结构

#数组的访问O(1)  链表的访问O(n)  数组可以通过索引查找，链表只能一个一个从头开始查找
#数组的插入或删除O(n)  链表的插入或删除O(1) 链表只需要修改地址的指向，数组是被修改位的后面都要挪位

#Definition for singly-linked list.
from typing import Optional, List
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        #定义一个占位节点，新的节点直接在该节点后插入
        dummy = ListNode(0)
        #tail始终指向最后一个节点
        tail = dummy
        #进位 0 / 1 ， 初始设置为0
        carry = 0

        p1 , p2 = l1 , l2
        while p1 != None or p2 != None or carry != 0 :
            #获取当前位的值
            x = p1.val if p1 != None else 0 #如果链条每结束获取当前节点的值，如果链条以结束0
            y = p2.val if p2 != None else 0

            #计算当前位
            sum = x + y + carry

            #留下个位
            digit = sum % 10
            #进位
            carry = sum // 10

            #个位作为新的节点插入
            tail.next = ListNode(digit)
            tail = tail.next  #尾节点后移

            if p1 != None:
                p1 = p1.next
            if p2 != None:
                p2 = p2.next
        return dummy.next
```

```python
def list2linklist(nums:List[int]) -> Optional[ListNode]:
    dummy = ListNode(0)
    curr = dummy
    for num in nums:
        curr.next = ListNode(num)
        curr = curr.next
    return dummy.next

def linkedlist_to_list(node: Optional[ListNode]) -> List[int]:
    """将链表转换为数组，方便观察结果"""
    result = []
    while node:
        result.append(node.val)
        node = node.next
    return result

if __name__ == "__main__":
    sol = Solution()

    l1 = list2linklist([2,4,3])
    l2 = list2linklist([5,6,4])
    result_node = sol.addTwoNumbers(l1,l2)
print(linkedlist_to_list(result_node))
```

```python
#递归法
from typing import Optional, List
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        total = l1.val + l2.val

        next1 = total // 10
        res = ListNode(total % 10)

        if l1.next is not None or l2.next is not None or next1 != 0 :
            if l1.next is not None :
                l1 = l1.next
            else:
                l1 = ListNode(0)
            if l2.next is not None :
                l2 = l2.next
            else:
                l2 = ListNode(0)

            l1.val = l1.val + next1
            res.next =  self.addTwoNumbers(l1,l2)


```

```python
class Solution:
    # 增加一个 carry 参数，默认为 0
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode], carry: int = 0) -> Optional[ListNode]:
        # 1. 递归出口：l1, l2 都走完了，且没有进位了
        if not l1 and not l2 and carry == 0:
            return None
        
        #2. 计算当前位的和
        v1 = l1.val if l1 else 0
        v2 = l2.val if l2 else 0
        total = v1 + v2 + carry

        #3. 创建结果节点
        res = ListNode(total % 10)
        new_carry = total // 10

        #4. 下层节点
        next_l1 = l1.next if l1 else None
        next_l2 = l2.next if l2 else None

        #递归
        res.next = self.addTwoNumbers(next_l1 , next_l2 , new_carry)
        return res
```
