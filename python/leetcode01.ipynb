```python
#1.暴力法
#用两次for循环，遍历数组，找到满足的下标返回
#为了避免重复，第二层循环从第一次循环的数的下一位开始循环
from typing import List
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        n = len(nums)
        for i in range(0,n,1):
            for j in range(i+1,n,1):
                if nums[i]+nums[j] == target:
                    return [i,j]

sol = Solution()
sol.twoSum([2,7,11,15],26)
```

```python
#2.哈希表法
#传入的列表其中所有内容都看作元素
#元素作为键 元素对应的索引作为值
#查哈希表，看看目标值-元素 是否存在于哈希表里 O(1)
from typing import List
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        map = {}
        for i in range(0,len(nums),1):
            map[nums[i]] = i
        for j in range(0,len(nums),1):
            diff  = target - nums[j]
            if diff in map and diff != nums[j]:
                return [map[diff],j]
solv = Solution()
solv.twoSum([3,2,4],6)


```

```python
#enumaerate(iterable,start = 0) 函数
#iterable 是可遍历对象如列表 元组 字符串 字典
#start 为下标起始值 默认从0开始
#返回一个对象（可以提取出下标 + 值的元组）
#在两数值和里面的应用如下
# for idx , num in enumerate(nums)

#重要思想：哈希表是在一边遍历列表的时候一边创建的
from typing import List
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        map = {} #创建一个空的哈希表 即字典
        for i ,num in enumerate(nums):
            diff = target - num #查看需要补的数 
            #1.如果需要补的数在哈希表里，那么之前已经遍历过这个补数了，直接返回两个索引
            if diff in map:
                return [map[diff],i]
            # 补数不存在，添加到哈希表里
            map[nums[i]] = i

sool = Solution()
sool.twoSum([3,2,4],7)
#sool.twoSum([3,2,3],6)

```
