## 下标地址
公式： 下标为i的地址 = 首地址 + i * 单个元素占用字节数

## 访问元素
访问数组中第index个元素  
1. 检查下标index是否满足， 0 ≤ index ≤ len(nums)-1
2. 符合：返回相应下标元素值
3. 不符合：抛出异常，返回特殊值

时间复杂度： O(1)

```python
def get_element(nums: list[int] , index: int):
    """
    获取数组指定下标元素值
    """
    if 0 <= index <= len(nums) - 1:
        return nums[index]
    else:
        raise IndexError(f"数组下标 {index}超出范围")

arr = [0 , 5 , 2 , 3 , 6 , 1 , 4]
print(get_element(arr , 5))
```

## 查找元素
1. 遍历数组，将目标值val与每个元素比较
2. 找到匹配元素时，返回其下标
3. 遍历完未找到时，返回特殊值 eg:-1

数组无序时，查找元素只能通过将val与数组中每个元素依次比较（线性查找）  
时间复杂度：O(n)

```python
def find_element(nums:list[int] , val: int):
    """查找元素值为val的位置"""
    for i in range(len(nums)):
        if nums[i] == val:
            return i
    return -1

arr = [0 , 5 , 6 , 3 , 7 , 1]
print(find_element(arr , 7))
```

## 插入元素
在数组第 index 个位置插入值 val
1. 检查 index 是否在 0 ≤ index ≤ len(nums)范围内
2. 扩展数组长度，给新元素腾出空间
3. 将index及其往后的元素整体向后移动一位
4. 在原index位置插入val  

移动元素的操作和元素个数有关，因此时间复杂度为O(n)


```python
def insert_element(nums:list[int] , index:int , val:int):
    """
    在指定位置插入元素
    """
    #检查index是否在有效范围内
    if 0 <= index <= len(nums):
        #扩展数组长度，在末尾添加一个占位元素
        nums.append(0)
        #将index及其后面的元素整体向后挪一位
        for i in range(len(nums) - 1 , index , -1): #倒序遍历，防止数组越界
            nums[i] = nums[i-1] 
        #在index位置插入val
        nums[index] = val
        return True
    else:
        #索引不在有效范围内
        return False

arr = [0 , 5 , 2 , 7 , 3 , 1 , 6]
result = insert_element(arr , 3 , 114514)
print(f"插入后数组: {arr}") 
```

## 改变元素
将数组中第index个元素值改为val
1. 检查 index 是否在 0 ≤ index ≤ len(nums) - 1
2. 将第 index 个元素值赋值为val

```python
def change_element(nums: list[int] , index: int , val: int):
    """修改数组中指定元素的元素值"""
    if 0 <= index < len(nums):
        nums[index] = val
        return True
    else:
        return False
    
arr = [0 , 5 , 1 , 7 , 3 , 9]
result = change_element(arr , 2 , 114514)
print(f"修改后的数组: {arr}")
```

## 删除元素
删除数组中第 index 个位置的元素
1. 检查下标是否在合法范围内， 0 ≤ index < len(nums)
2. 将index + 1及其后面的元素整体向前挪一位
3. 更新数组长度，或者删除最后一个元素

```python
def delete_element(nums: list[int] , index: int):
    """删除数组中指定位置的元素"""
    if 0 <= index < len(nums):
        #将 index 后的元素整体向前挪动一位
        for i in range(index , len(nums) - 1 , 1):
            nums[i] = nums[i+1]
        #删除最后一个元素
        nums.pop()
        return True
    else:
        return False
    
arr = [0, 5, 2, 3, 7, 1, 6]
result = delete_element(arr , 6)
print(f"删除后数组: {arr}")
```
