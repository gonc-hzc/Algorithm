## 排序算法
常见的排序算法：
1. 冒泡排序：通过相邻元素的比较、交换，将最大元素逐步冒泡排到数组末尾
2. 选择排序：每次在未排序区间选择最小元素放到已排序区间末尾
3. 插入排序：将未排序区间的元素插入到已排序区间的合适位置
4. 希尔排序：先按一定间隔分组进行插入排序，逐步缩小间隔，最后整体插入排序
5. 归并排序：将数组分成两半，分别排序后合并
6. 快速排序：选择一个基准元素，将数组分成两部分，递归排序
7. 堆排序：利用堆这种数据结构进行排序
8. 计数排序：统计每个元素出现的次数，按顺序输出
9. 通排序：将元素分到有限数量的桶中，对每个桶单独排序
10. 基数排序：按照元素的位数排序

## 冒泡排序
对于一个长度为 n 的数组，冒泡排序步骤如下：
1. 第一轮冒泡：对前n个元素，依次比较相邻元素，较大的元素向右交换，知道最大值移动到数组末尾.
    1. 比较第1个和第2个元素，若前者大于后者，交换两个元素位置
    2. 比较第2个和第3个元素，若前者大于后者，交换两个元素位置
    3. 依次类推，依次比较相邻元素，直到第n-1个和第n个元素
    4. 第一轮最大元素已经排到数组末尾
2. 第二轮冒泡：对前n-1个元素进行如上操作
    1. 比较第1个和第2个元素，如果前者大于后者则交换
    2. 比较第2个和第3个元素，如果前者大于后者则交换
    3. 直到比较第n−2个和第n−1个元素
    4. 完成后第2大的元素已经排到正确位置了
3. 重复进行上述操作，每完成一轮，需要比较的个数 -1 

到了只剩下1个元素时，这个元素已经在正确的位置，因此总共有 n - 1 轮排序。
在第 i 轮排序中，内部需要 n - i - 1 次比较

```python
"""冒泡排序实现"""
class Solution:
    def bubbleSort(self , nums: list[int]) -> list[int]:
        n = len(nums)

        #外层循环控制轮数：每一轮将当前未排序区间中的最大值移到未排序区间末尾
        for i in range(n-1): #从0开始到n-2结束，共n-1轮
            swapped = False #记录本趟是否发生过交换
            #内层循环：相邻元素两两比较，较大值后移
            for j in range(n-i-1):
                #较大值后移实现
                if nums[j] > nums[j+1]:
                    nums[j] , nums[j+1] = nums[j+1] , nums[j]
                    swapped = True #发生了交换
            #若本轮中没有发生任何交换，说明数组已经是有序的，可以提前结束
            if not swapped:
                break
        return nums #返回排序后的数组
    
    def sortArray(self , nums):
        """冒泡排序的接口"""
        return self.bubbleSort(nums)
    
arr = [3 , 9 , 5 , 6 , 1 , 0 , 7 ,54]
sol = Solution() #创建对象
res = sol.sortArray(arr)
print(f"排序后的结果: {res}")

```

## 选择排序
将数组分为两个区间：左侧为已排序区间，右侧为未排序区间
每趟从未排序区间中选择最小的元素，放到已排序区间的末尾（等同未排序区间的第一个元素）

选择排序步骤：
假设有一个长度为n的数组
1. 初始状态：已排序区间为空，未排序空间为[0 , n-1]
2. 第i趟选择（i从1开始）：
    1. 在未排序区间[i-1 , n-1]中找到最小元素位置min_i
    2. 将位置 i - 1的元素与位置min_i元素互换
    3. 此时[0 , i-1]为已排序空间，[i , n-1]为未排序空间
3. 重复步骤2 直到未排序空间为0


```python
"""选择排序实现"""
class Solution:
    def selectionSort(self , nums:list[int])->list[int]:
        n = len(nums)
        for i in range(n-1):
            #找到未排序区间中最小元素的位置
            min_i = i
            for j in range(i+1 , n):
                if nums[j] < nums[min_i]:
                    min_i = j
            #将最小元素与未排序区间的第一个元素交换
            nums[i] , nums[min_i] = nums[min_i] , nums[i]
        return nums #返回排序后的数组
    
    def sortArray(self , nums: list[int])->list[int]:
        """选择排序的接口"""
        return self.selectionSort(nums)
    
arr = [3,46,15,6,8,26,17,7,32,1]
sol = Solution() #创建对象
res = sol.sortArray(arr)
print(arr)
```

## 插入排序
将数组分为有序区间和无序区间，每次从无序区间中取出一个元素插入到有序区间的正确位置

插入排序算法步骤：
1. 初始化：有序空间为[0 , 0] , 无序空间为[1 , n-1]
2. 第 i 趟插入（i 从 1 到 n-1）
    1. 取出无序区间第一个元素 nums[i]
    2. 从右到左遍历有序区间，将大于 nums[i] 的元素右移一位
    3. 找到合适的位置，插入nums[i]
    4. 有序区间扩展为 [0 , i] , 无序区间变为 [i+1 , n-1]

```python
"""插入排序实现"""
class Solution:
    def insertSort(self , nums:list[int]) ->list[int]:
        #遍历无序区间
        for i in range(1 , len(nums)):
            key = nums[i] #当前要插入的元素
            j = i  # j为无序区间的起始位置

            #从右到左遍历有序区间，找到key的正确位置插入
            while j > 0 and nums[j-1] > key:
                nums[j] = nums[j-1] #将较大的元素后移
                j -= 1
            nums[j] = key #将key插入到正确位置
        return nums #返回排序后的数组

    def sortArray(self , nums:list[int]) -> list[int]:
        """插入排序的接口"""
        return self.insertSort(nums)

arr = [3,46,15,6,8,26,17,7,32,1]
sol = Solution()
res = sol.sortArray(arr)
print(arr)
```

## 希尔排序
通过设定不同间隔(gap)，将数组分组进行插入排序，然后逐步缩小间隔直至为1，最终完成整个数组的排序

普通插入排序的问题是：比如一个数组[9,6,7,6 ... ,2,1]，1要挪到正确位置需要移动很多次

希尔排序算法步骤：  假设数组有n个元素
1. 设定初始间隔 gap = n / 2
2. 按间隔将数组分组，对每组进行插入排序
3. 缩小间隔 gap = gap // 2
4. 重复步骤2、3 ， 直到gap = 1
5. 最后对整个数组进行一次插入排序

```python
"""希尔排序实现"""
class Solution:
    def shellSort(self , nums:list[int])->list[int]:
        size = len(nums)
        gap = size // 2 #初始增量

        #不断缩小gap , 直到gap为0
        while gap > 0:
            #对每个gap分组进行插入排序
            for i in range(gap , size):
                key = nums[i] #当前要插入的元素
                j = i # j为当前元素在分组中的位置

                #从右到左遍历分组中的有序区间，找到key的正确位置插入
                while j >= gap and nums[j-gap] > key:
                    nums[j] = nums[j-gap]
                    j -= gap
                nums[j] = key #将key插入到正确位置

            gap //= 2 #缩小gap` 
            
        return nums #返回排序后的数组
    def sortArray(self , nums:list[int])->list[int]:
        """希尔排序接口"""
        return self.shellSort(nums)

arr = [7,2,6,8,0,4,1,5,9,3]
sol = Solution()
res = sol.sortArray(arr)
print(res)
print(arr)
```

## 归并排序
利用分治法，将数组递归地一分为二，直至每个子数组只包含一个元素，随后将有序地子数组两两合并，最终得到一个整体有序地数组

归并算法步骤： 数组元素个数 n 个
1. 分解过程：递归地将当前数组分为两部分，直至子数组只包含一个元素为止
    1. 找到数组的中间位置mid，将数组划分为左、右两个子数组 left_nums 和 right_nums
    2. 分别对left_nums 和 right_nums递归执行分解操作
    3. 最终将原数组拆分为 n 个长度为 1 的有序子数组 
2. 归并过程：从长度为1的有序子数组开始，逐步将相邻的有序子数组两两合并，最终合并成一个长度为 n 的有序数组
    1. 新建数组 nums 用于存放合并后的有序结果
    2. 设置两个指针 left_i , right_i ， 分别指向left_nums 和 right_nums起始位置
    3. 比较两个指针所指元素。将较小者加入结果数组 nums , 并将对应指针后移一位
    4. 重复上述操作，直到某一指针到达对应子数组末尾
    5. 将另一个子数组剩余的所有元素依次加入结果数组 nums
    6. 返回合并后的有序数组 nums

```python
"""归并排序实现"""
class Solution:
    def merge(self , left_nums:list[int] , right_nums:list[int])->list[int]:
        nums = []
        left_i , right_j = 0 , 0 #分别为左右数组的指针

        # 合并两个有序子数组
        while left_i < len(left_nums) and right_j < len(right_nums):
            if left_nums[left_i] < right_nums[right_j]:
                nums.append(left_nums[left_i])
                left_i += 1
            else:
                nums.append(right_nums[right_j])
                right_j += 1
        #将剩余元素添加到结果数组中
        while left_i < len(left_nums):
            nums. append(left_nums[left_i])
            left_i += 1
        while right_j < len(right_nums):
            nums.append(right_nums[right_j])
            right_j += 1
        
        return nums #返回合并后的有序数组

    def mergeSort(self , nums:list[int])->list[int]:

        #递归终止条件：当数组长度为0或1时，已经是有序的
        if len(nums) <= 1:
            return nums
        
        mid = len(nums) // 2 #找到数组的中点
        left_nums = self.mergeSort(nums[0:mid]) #递归分解左半部分
        right_nums = self.mergeSort(nums[mid:]) #递归分解右半部分

        return self.merge(left_nums, right_nums) #合并两个有序子数组

    def sortArray(self , nums:list[int])->list[int]:
        """归并排序接口"""
        return self.mergeSort(nums)

arr = [7,2,6,8,0,4,1,5,9,3]
sol = Solution()
res = sol.sortArray(arr)
print(res)
```
